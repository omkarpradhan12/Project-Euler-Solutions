```python
#Completed
'''
1)"If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.
Find the sum of all the multiples of 3 or 5 below 1000."
'''
def multi():
    x = []
    for i in range(0,1000):
        if i%3==0 or i%5==0:
            x.append(i)
    return sum(x)

multi()
```

```python
#Completed
'''
3)"The prime factors of 13195 are 5, 7, 13 and 29.
What is the largest prime factor of the number 600851475143 ?"

'''


def prime_check(n):
    #using fernats little theorem
    return ((2**n)-2)%n

def largest_factor(n):
    lf = 0
    for i in range(2,n):
        if (n%i==0 and prime_check(i)==0):
            lf=i
            
    return lf



largest_factor(600851475143)
```

```python
#Completed
'''
6)"The sum of the squares of the first ten natural numbers is,
12 + 22 + ... + 102 = 385
The square of the sum of the first ten natural numbers is,
(1 + 2 + ... + 10)2 = 552 = 3025
Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 − 385 = 2640.
Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum."

'''

def diff():
    sums_sq = 0
    sq_sum = 0
    for i in range(1,101):
        sums_sq+=(i*i)
        sq_sum+=i
    return (sq_sum**2)-sums_sq

diff()
```

```python
#Completed
'''
"A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
a2 + b2 = c2
For example, 32 + 42 = 9 + 16 = 25 = 52.
There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc."

'''

possible_trips = []
for i in range(1,999):
    for j in range(1,1000-i):
        if sum([i,1000-i-j,j])==1000:
            if i<(1000-i-j):
                if (1000-i-j)<j:
                    possible_trips.append((i,1000-i-j,j))

for i in possible_trips:
    a = i[0]
    b = i[1]
    c = i[2]
    
    if (a**2 + b**2)==c**2:
        print(a,b,c,a*b*c)

```

```python
#Completed
'''
"The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.
Find the sum of all the primes below two million."
'''
def prime_checker(n):
    if n==1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True
        
    


def sum_prime(n):
    sumofprimes = 0
    for i in range(2,n):
        if prime_checker(i):
            sumofprimes+=i
    return sumofprimes

sum_prime(2000000)
```
